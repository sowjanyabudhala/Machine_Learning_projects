**Oversampling**:done by using SMOTE(Synthetic minority oversampling technique).It synthesizes new minoriy instances between existing  minority instances.smote generates virtual training records by interpolation.
Advantage of smote is not generate duplicates but it creates synthetic datapoints.

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
X.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res=SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.2,random_state=42)
```
